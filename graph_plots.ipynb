```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.get_option("display.max_columns",50)
pd.set_option('display.max_rows', 50)
```

```python
job_df = pd.read_json("processed_naukri_job_data.json")
```

```python
sns.set_style('darkgrid')
```

```python
job_df["location"].unique()
```

```python
metro_cities=["Mumbai","Pune","Bangalore","Noida","Hyderabad","Chennai","Thane","Ahmedabad"]
```

```python
job_metro_cities=job_df[job_df["location"].isin(metro_cities)]
job_metro_cities.head()
```

```python
sns.barplot(x ='location', y ='average_salary', data = job_metro_cities)
plt.xticks(rotation=90)
plt.show()
```
